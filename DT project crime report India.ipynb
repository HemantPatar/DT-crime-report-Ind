```python
KNA_SG = KNA[KNA['Sub_Group_Name']== '14. Total (Sum of 1-13 Above)']  #keeping only total category of subgroup
K_A = KNA_SG.groupby(['Year'])['K_A_Cases_Reported'].sum().reset_index() #grouping
sns.set_context("talk")
plt.style.use("fivethirtyeight")
plt.figure(figsize = (14,10))
#sns.palplot(sns.color_palette("hls", 8))
ax = sns.barplot(x = 'Year' , y = 'K_A_Cases_Reported', data = K_A ,palette= 'dark') #plotting bar graph
plt.title("Total reported cases of kidnapping per Year")
ax.set_ylabel('')
for p in ax.patches:
             ax.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=15, color='black', xytext=(0, 8),
                 textcoords='offset points')
```

```python
KNA_SG = KNA[KNA['Sub_Group_Name']== '14. Total (Sum of 1-13 Above)']  #keeping only total category of subgroup
K_A = KNA_SG.groupby(['Year'])['K_A_Grand_Total'].sum().reset_index() #grouping
sns.set_context("talk")
plt.style.use("fivethirtyeight")
plt.figure(figsize = (14,10))
#sns.palplot(sns.color_palette("hls", 8))
ax = sns.barplot(x = 'Year' , y = 'K_A_Grand_Total', data = K_A ,palette= 'dark') #plotting bar graph
plt.title("Total Victims of kidnapping per Year")
ax.set_ylabel('')
for p in ax.patches:
             ax.annotate("%.f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=15, color='black', xytext=(0, 8),
                 textcoords='offset points')
```

```python
k_A_a = KNA.groupby(['Year'])['K_A_Female_10_15_Years',
       'K_A_Female_15_18_Years', 'K_A_Female_18_30_Years',
       'K_A_Female_30_50_Years', 'K_A_Female_Above_50_Years',
        'K_A_Female_Upto_10_Years'].sum().reset_index()  #grouby year and age group
k_A_a = k_A_a.melt('Year', var_name='AgeGroup',  value_name='vals') #melting the dataset

plt.style.use("fivethirtyeight")
plt.figure(figsize = (14,10))
ax = sns.barplot(x = 'Year' , y = 'vals',hue = 'AgeGroup' ,data = k_A_a ,palette= 'bright') #plotting a bar
plt.title('Age Distribution of female Victims per Year',size = 20)
ax.get_legend().set_bbox_to_anchor((1, 1)) #anchoring the labels so that they dont show up on the graph
ax.set_ylabel('')
```

```python
k_A_a = KNA.groupby(['Year'])[
       'K_A_Male_10_15_Years', 'K_A_Male_15_18_Years', 'K_A_Male_18_30_Years',
       'K_A_Male_30_50_Years', 'K_A_Male_Above_50_Years', 
       'K_A_Male_Upto_10_Years'].sum().reset_index()  #grouby year and age group
k_A_a = k_A_a.melt('Year', var_name='AgeGroup',  value_name='vals') #melting the dataset

plt.style.use("fivethirtyeight")
plt.figure(figsize = (14,10))
ax = sns.barplot(x = 'Year' , y = 'vals',hue = 'AgeGroup' ,data = k_A_a ,palette= 'bright') #plotting a bar
plt.title('Age Distribution of male Victims per Year',size = 20)
ax.get_legend().set_bbox_to_anchor((1, 1)) #anchoring the labels so that they dont show up on the graph
ax.set_ylabel('')
```

```python
# let's plot the unreported rape cases sorted by states throughout 2001 to 2010
victims_by_state = KNA.groupby('Area_Name').sum()
victims_by_state.drop('Year', axis = 1, inplace = True)

# let's finally plot it
plt.subplots(figsize = (15, 6))
ct = victims_by_state[victims_by_state['K_A_Grand_Total'] 
                                 > 0]['K_A_Grand_Total'].sort_values(ascending = False)
#print(ct)
ax = ct.plot.bar()
ax.set_xlabel('Area Name')
ax.set_ylabel('Total Number of Victims from 2001 to 2010')
ax.set_title('Statewise total Victims throughout 2001 to 2010')
plt.show()
```

```python
# let's plot the unreported rape cases sorted by states throughout 2001 to 2010
victims_by_state = KNA.groupby('Area_Name').sum()
victims_by_state.drop('Year', axis = 1, inplace = True)

# let's finally plot it
plt.subplots(figsize = (15, 6))
ct = victims_by_state[victims_by_state['K_A_Cases_Reported'] 
                                 > 0]['K_A_Cases_Reported'].sort_values(ascending = False)
#print(ct)
ax = ct.plot.bar()
ax.set_xlabel('Area Name')
ax.set_ylabel('Total Number of reported cases from 2001 to 2010')
ax.set_title('Statewise total reported cases throughout 2001 to 2010')
plt.show()
```
